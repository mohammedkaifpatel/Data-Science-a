```python
import numpy as np
import pandas as pd
```

```python
titanic_train=pd.read_csv(r"D:\Techno Hacks\titanic CleanData.csv")
```

```python
titanic_train
```

```python
# Display summary statistics for numerical columns
summary_stats = titanic_train.describe()
summary_stats
```

```python
# Display additional statistics for non-numerical columns
non_numeric_stats = titanic_train.describe(include='all')
print("\nSummary Statistics for All Columns:")
print(non_numeric_stats)
```

```python
selected_column = 'Age'

# Calculate mean
mean_value = titanic_train[selected_column].mean()

# Calculate median
median_value = titanic_train[selected_column].median()

# Calculate mode (Note: Mode may have multiple values, so it returns a Series)
mode_values = titanic_train[selected_column].mode()

# Calculate standard deviation
std_deviation_value = titanic_train[selected_column].std()

# Print the results
print(f"Mean of {selected_column}: {mean_value}")
print(f"Median of {selected_column}: {median_value}")
print(f"Mode of {selected_column}: {mode_values.to_list()}")
print(f"Standard Deviation of {selected_column}: {std_deviation_value}")
```

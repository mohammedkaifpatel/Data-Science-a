```python
# Fill missing values in the 'Embarked' column with the mode
most_common_embarked = titanic_train['Embarked'].mode()[0]
titanic_train['Embarked'].fillna(most_common_embarked, inplace=True)

# Verify that null values have been handled
print(titanic_train['Embarked'].isnull().sum())
```

```python
# Calculate the mode of the 'Cabin' column, excluding NaN values
cabin_mode =titanic_train['Cabin'].mode()

# Fill missing values in the 'Cabin' column with the mode
titanic_train['Cabin'].fillna(cabin_mode[0] if not cabin_mode.empty else "No mode", inplace=True)
```

```python
# Calculate the median age
median_age = titanic_train['Age'].median()

# Fill missing values in the 'Age' column with the median age
titanic_train['Age'].fillna(median_age, inplace=True)

# Verify that null values have been handled
print(titanic_train['Age'].isnull().sum())
```
